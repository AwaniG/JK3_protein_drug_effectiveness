```python
 #Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
X_train = pd.read_csv('/content/X_train.csv')
y_train = pd.read_csv('/content/y_train.csv')
X_test = pd.read_csv('/content/X_test.csv')
y_test = pd.read_csv('/content/y_test.csv')
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## MLP

```python
import numpy as np
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the scaler on the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
regressor = MLPRegressor(hidden_layer_sizes=(50,50), max_iter=1000, random_state=42, learning_rate_init=0.01)

# Perform 5-fold cross-validation
mse_scores = cross_val_score(regressor, X_train_scaled, y_train.values.ravel(), cv=5, scoring='neg_mean_squared_error')
avg_mse = -mse_scores.mean()

print(f"5-Fold Cross-Validation MSE: {avg_mse:.5f}")

# Train the regressor on the full training set
regressor.fit(X_train_scaled, y_train.values.ravel())


# Predict on the test set
y_pred_test = regressor.predict(X_test_scaled)

# Calculate test MSE
mse_test = mean_squared_error(y_test.values.ravel(), y_pred_test)
print(f"Test MSE: {mse_test:.5f}")
```
