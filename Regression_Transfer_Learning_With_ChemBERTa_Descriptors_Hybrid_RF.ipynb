```python
# Load embedded datasets
train_embed_zinc_df = pd.read_csv('/content/drive/Shareddrives/1:1 Awani Gadre/Dataset/train_chemberta_descriptors_zinc.csv')
test_embed_zinc_df = pd.read_csv ('/content/drive/Shareddrives/1:1 Awani Gadre/Dataset/test_chemberta_descriptors_zinc.csv')
train_embed_zinc_df.dropna(inplace=True)
test_embed_zinc_df.dropna(inplace=True)
```

```python
X_train = train_embed_zinc_df.drop(columns=['pIC50'])
y_train = train_embed_zinc_df['pIC50']

X_test = test_embed_zinc_df.drop(columns=['pIC50'])
y_test = test_embed_zinc_df['pIC50']

# from sklearn.neural_network import MLPRegressor

# regressor = MLPRegressor(hidden_layer_sizes=(500,100), max_iter=1000, random_state=42, learning_rate_init=0.0001)

# # Perform 5-fold cross-validation
# mse_scores = cross_val_score(regressor, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
# avg_mse = -mse_scores.mean()

# print(f"5-Fold Cross-Validation MSE: {avg_mse:.5f}")

# # Train the regressor on the full training set
# regressor.fit(X_train, y_train)

# # Predict on the test set
# y_pred_test = regressor.predict(X_test)

# # Calculate test MSE
# mse_test = mean_squared_error(y_test, y_pred_test)
# print(f"Test MSE: {mse_test:.5f}")

from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=500, random_state=42)

# Perform 5-fold cross-validation
mse_scores = cross_val_score(rf_regressor, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
avg_mse = -mse_scores.mean()

print(f"5-Fold Cross-Validation MSE: {avg_mse:.5f}")

# Train the regressor on the full training set
rf_regressor.fit(X_train, y_train)

# Predict on the test set
y_pred_test = rf_regressor.predict(X_test)

# Calculate test MSE
mse_test = mean_squared_error(y_test, y_pred_test)
print(f"Test MSE: {mse_test:.5f}")
```
